````python
import os
import re
import gc
import cv2
import json
import time
import math
import random
import hashlib
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET

from pathlib import Path
from collections import defaultdict
from tqdm.auto import tqdm
from skimage.measure import shannon_entropy
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
    roc_curve,
    precision_recall_curve,
)

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import models

warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
cv2.setNumThreads(0)

DATASET_ROOT = Path("/kaggle/input/datasets/angelchi56/abnormal-highdensity-crowds/Abnormal High-density Crowds")
OUTPUT_DIR = Path("/kaggle/working/crowd_anomaly_deep_project")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}
VID_EXTS = {".mp4", ".avi", ".mov", ".mkv", ".webm", ".mpeg", ".mpg"}

SEQ_LEN = 12
SEQ_STRIDE = 2
SEQ_SAMPLE_STEP = 8
MAX_SEQS_TRAIN = 120
MAX_SEQS_EVAL = 80
IMG_SIZE = (192, 192)
FEATURE_IMG_SIZE = (128, 128)
TRAIN_SPLIT = 0.70
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15
BATCH_SIZE = 4
NUM_EPOCHS = 8
LEARNING_RATE = 2e-4
WEIGHT_DECAY = 1e-4
EARLY_STOPPING = 3
THRESHOLDS = np.linspace(0.10, 0.90, 81)
NUM_WORKERS = 2
PIN_MEMORY = True

NORM_MEAN = np.array([0.485, 0.456, 0.406], dtype=np.float32)
NORM_STD = np.array([0.229, 0.224, 0.225], dtype=np.float32)


def natural_key(text):
    return [int(chunk) if chunk.isdigit() else chunk.lower() for chunk in re.split(r"(\d+)", str(text))]


def tokenize_path(path_like):
    return [tok for tok in re.split(r"[^a-z0-9]+", str(path_like).lower()) if tok]


def infer_scene(path_like):
    path_obj = Path(path_like)
    for part in path_obj.parts:
        if re.match(r"^\d+_", part.lower()):
            return part
    return path_obj.parent.name if path_obj.parent.name else "unknown_scene"


def infer_label(path_like):
    tokens = set(tokenize_path(path_like))
    joined = " ".join(tokens)
    if "italy" in tokens or ("4" in tokens and "italy" in joined):
        return 1
    if "normal" in tokens or "train" in tokens:
        return 0
    if "abnormal" in tokens or "test" in tokens:
        return 1
    return None


def infer_role(path_like):
    tokens = set(tokenize_path(path_like))
    if "train" in tokens:
        return "train"
    if "val" in tokens or "valid" in tokens or "validation" in tokens:
        return "val"
    if "test" in tokens:
        return "test"
    return "unspecified"


def safe_float(value, default=np.nan):
    try:
        return float(value)
    except Exception:
        return default


def collect_clips(root):
    clips = []
    frame_dirs = defaultdict(list)

    for path in root.rglob("*"):
        if not path.is_file():
            continue
        suffix = path.suffix.lower()
        if suffix in VID_EXTS:
            cap = cv2.VideoCapture(str(path))
            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)
            fps = safe_float(cap.get(cv2.CAP_PROP_FPS), default=np.nan)
            cap.release()
            rel = path.relative_to(root)
            clips.append(
                {
                    "source_type": "video",
                    "source_path": str(path),
                    "video_id": rel.with_suffix("").as_posix(),
                    "scene": infer_scene(rel),
                    "label": infer_label(rel),
                    "predefined_role": infer_role(rel),
                    "num_frames": frame_count,
                    "fps": fps,
                }
            )
        elif suffix in IMG_EXTS:
            frame_dirs[path.parent].append(path)

    for frame_dir, files in frame_dirs.items():
        rel = frame_dir.relative_to(root)
        clips.append(
            {
                "source_type": "frames",
                "source_path": str(frame_dir),
                "video_id": rel.as_posix(),
                "scene": infer_scene(rel),
                "label": infer_label(rel),
                "predefined_role": infer_role(rel),
                "num_frames": len(files),
                "fps": np.nan,
            }
        )

    clips_df = pd.DataFrame(clips).drop_duplicates(subset=["video_id", "source_type"]).reset_index(drop=True)
    if clips_df.empty:
        raise FileNotFoundError(f"No videos or frame folders found under: {root}")
    return clips_df


def find_xml_for_clip(source_path):
    source = Path(source_path)
    roots = []
    if source.is_dir():
        roots.extend([source, source.parent, source.parent.parent if source.parent != source else source])
    else:
        roots.extend([source.parent, source.parent.parent if source.parent != source else source.parent])

    best_match = None
    best_score = -1
    source_tokens = set(tokenize_path(source.stem if source.is_file() else source.name))

    for root in roots:
        if not root or not root.exists():
            continue
        for xml_path in root.glob("*.xml"):
            score = len(source_tokens.intersection(set(tokenize_path(xml_path.stem))))
            if score > best_score:
                best_match = xml_path
                best_score = score

    scene_dir = DATASET_ROOT / infer_scene(source_path)
    if scene_dir.exists():
        for xml_path in scene_dir.rglob("*.xml"):
            score = len(source_tokens.intersection(set(tokenize_path(xml_path.stem))))
            if score > best_score:
                best_match = xml_path
                best_score = score

    return str(best_match) if best_match is not None else None


def parse_xml_abnormal_frames(xml_path):
    abnormal_frames = {}
    if not xml_path:
        return abnormal_frames

    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()
    except Exception as exc:
        print(f"Skipping XML {xml_path}: {exc}")
        return abnormal_frames

    for elem in root.iter():
        tag = elem.tag.lower().split("}")[-1]
        attrs = {k.lower(): str(v).lower() for k, v in elem.attrib.items()}

        frame_idx = None
        for key in ("frame", "frameid", "number", "id", "index"):
            if key in attrs:
                match = re.search(r"\d+", attrs[key])
                if match:
                    frame_idx = int(match.group())
                    break

        if frame_idx is None:
            continue

        payload = " ".join([tag, attrs.get("label", ""), attrs.get("name", ""), attrs.get("class", ""), elem.text or ""]).lower()
        if tag in {"box", "polygon", "mask", "object", "track", "points"}:
            abnormal_frames[frame_idx] = 1
        elif any(word in payload for word in ("abnormal", "anomaly", "panic", "incident", "crowd")):
            abnormal_frames[frame_idx] = 1

    return abnormal_frames


def make_video_splits(clips_df):
    videos = clips_df[["video_id", "label"]].drop_duplicates().reset_index(drop=True)
    if len(videos) < 3:
        raise RuntimeError("Not enough unique clips to create train/validation/test splits.")

    can_stratify = videos["label"].nunique() > 1 and videos["label"].value_counts().min() >= 2
    for seed in range(SEED, SEED + 100):
        try:
            train_videos, temp_videos = train_test_split(
                videos,
                test_size=(1.0 - TRAIN_SPLIT),
                random_state=seed,
                stratify=videos["label"] if can_stratify else None,
            )
            temp_can_stratify = temp_videos["label"].nunique() > 1 and temp_videos["label"].value_counts().min() >= 2
            val_videos, test_videos = train_test_split(
                temp_videos,
                test_size=(TEST_SPLIT / (VAL_SPLIT + TEST_SPLIT)),
                random_state=seed,
                stratify=temp_videos["label"] if temp_can_stratify else None,
            )
            return set(train_videos["video_id"]), set(val_videos["video_id"]), set(test_videos["video_id"])
        except ValueError:
            continue

    shuffled = videos.sample(frac=1.0, random_state=SEED).reset_index(drop=True)
    n_total = len(shuffled)
    n_train = max(1, int(round(n_total * TRAIN_SPLIT)))
    n_val = max(1, int(round(n_total * VAL_SPLIT)))
    train_ids = set(shuffled.iloc[:n_train]["video_id"])
    val_ids = set(shuffled.iloc[n_train : n_train + n_val]["video_id"])
    test_ids = set(shuffled.iloc[n_train + n_val :]["video_id"])
    return train_ids, val_ids, test_ids


def get_frame_file_cache(clips_df):
    cache = {}
    for _, clip in clips_df[clips_df["source_type"] == "frames"].iterrows():
        frame_dir = Path(clip["source_path"])
        cache[clip["video_id"]] = sorted(
            [p for p in frame_dir.iterdir() if p.is_file() and p.suffix.lower() in IMG_EXTS],
            key=natural_key,
        )
    return cache


def get_sequence_indices(total_frames, seq_len=SEQ_LEN, seq_stride=SEQ_STRIDE, sample_step=SEQ_SAMPLE_STEP, max_sequences=MAX_SEQS_EVAL):
    required = (seq_len - 1) * seq_stride + 1
    if total_frames < required:
        return []

    starts = list(range(0, total_frames - required + 1, sample_step))
    if not starts:
        starts = [0]
    if max_sequences and len(starts) > max_sequences:
        keep = np.linspace(0, len(starts) - 1, max_sequences, dtype=int)
        starts = [starts[idx] for idx in keep]
    return [list(range(start, start + required, seq_stride)) for start in starts]


def sequence_label(indices, clip_label, xml_frame_map):
    if not xml_frame_map:
        return int(clip_label)
    abnormal_votes = sum(int(xml_frame_map.get(idx, xml_frame_map.get(idx + 1, 0))) for idx in indices)
    return int(abnormal_votes > 0)


def build_sequence_manifest(clips_df):
    frame_cache = get_frame_file_cache(clips_df)
    manifest_rows = []

    for _, clip in clips_df.iterrows():
        xml_map = parse_xml_abnormal_frames(clip["xml_path"]) if pd.notna(clip["xml_path"]) else {}
        max_sequences = MAX_SEQS_TRAIN if clip["split"] == "train" else MAX_SEQS_EVAL
        sequences = get_sequence_indices(
            total_frames=int(clip["num_frames"]),
            max_sequences=max_sequences,
        )
        for seq_num, indices in enumerate(sequences):
            manifest_rows.append(
                {
                    "sequence_id": f"{clip['video_id']}__seq_{seq_num:04d}",
                    "video_id": clip["video_id"],
                    "scene": clip["scene"],
                    "source_type": clip["source_type"],
                    "source_path": clip["source_path"],
                    "split": clip["split"],
                    "label": sequence_label(indices, clip["label"], xml_map),
                    "frame_indices": json.dumps(indices),
                    "start_frame": int(indices[0]),
                    "end_frame": int(indices[-1]),
                    "center_frame": int(indices[len(indices) // 2]),
                    "num_frames_video": int(clip["num_frames"]),
                    "fps": clip["fps"],
                    "xml_path": clip["xml_path"],
                }
            )

    manifest = pd.DataFrame(manifest_rows)
    if manifest.empty:
        raise RuntimeError("No frame sequences could be created. Try reducing SEQ_LEN or SEQ_STRIDE.")

    return manifest, frame_cache


def load_video_sequence(video_path, frame_indices):
    cap = cv2.VideoCapture(str(video_path))
    frames = []
    current_target_idx = 0
    target_frames = sorted(frame_indices)
    if not target_frames:
        cap.release()
        return frames

    cap.set(cv2.CAP_PROP_POS_FRAMES, target_frames[0])
    current_frame = target_frames[0]

    while current_target_idx < len(target_frames):
        ok, frame = cap.read()
        if not ok:
            break
        if current_frame == target_frames[current_target_idx]:
            frames.append(frame)
            current_target_idx += 1
        current_frame += 1

    cap.release()
    return frames


def load_frame_sequence(frame_files, frame_indices):
    frames = []
    for idx in frame_indices:
        if idx >= len(frame_files):
            continue
        frame = cv2.imread(str(frame_files[idx]))
        if frame is not None:
            frames.append(frame)
    return frames


def stable_seed(text):
    return int(hashlib.md5(str(text).encode("utf-8")).hexdigest()[:8], 16)


def augment_sequence(frames, seed_value):
    rng = np.random.default_rng(seed_value)
    apply_flip = rng.random() < 0.5
    brightness = float(1.0 + rng.uniform(-0.18, 0.18))
    bias = float(rng.uniform(-20, 20))
    blur = rng.random() < 0.25
    noise_std = float(rng.uniform(3, 8)) if rng.random() < 0.25 else 0.0

    aug_frames = []
    for frame in frames:
        img = frame.copy()
        if apply_flip:
            img = cv2.flip(img, 1)
        img = cv2.convertScaleAbs(img, alpha=brightness, beta=bias)
        if blur:
            img = cv2.GaussianBlur(img, (3, 3), 0)
        if noise_std > 0:
            noise = rng.normal(0, noise_std, img.shape).astype(np.float32)
            img = np.clip(img.astype(np.float32) + noise, 0, 255).astype(np.uint8)
        aug_frames.append(img)
    return aug_frames


def flow_histogram(flow, bins=8):
    magnitude, angle = cv2.cartToPolar(flow[..., 0], flow[..., 1], angleInDegrees=False)
    hist, _ = np.histogram(angle, bins=bins, range=(0, 2 * np.pi), weights=magnitude)
    hist = hist.astype(np.float32)
    hist /= hist.sum() + 1e-6
    return magnitude, hist


HANDCRAFTED_FEATURES = [
    "intensity_mean_avg",
    "intensity_std_avg",
    "entropy_avg",
    "edge_density_avg",
    "laplacian_var_avg",
    "foreground_ratio_avg",
    "crowd_proxy_avg",
    "temporal_diff_mean_avg",
    "temporal_diff_std_avg",
    "flow_mean_avg",
    "flow_std_avg",
    "flow_q90_avg",
    "flow_max_avg",
    "brightness_drift",
    "motion_peak",
    "abrupt_change_ratio",
] + [f"flow_dir_bin_{i}" for i in range(8)]


def compute_handcrafted_features(frames):
    bg_sub = cv2.createBackgroundSubtractorMOG2(history=32, varThreshold=16, detectShadows=False)
    intensity_mean = []
    intensity_std = []
    entropy_vals = []
    edge_density = []
    lap_var = []
    foreground_ratio = []
    crowd_proxy = []
    temporal_diff_mean = []
    temporal_diff_std = []
    flow_mean = []
    flow_std = []
    flow_q90 = []
    flow_max = []
    flow_hist_total = np.zeros(8, dtype=np.float32)

    prev_gray = None
    for frame in frames:
        frame_small = cv2.resize(frame, FEATURE_IMG_SIZE, interpolation=cv2.INTER_AREA)
        gray = cv2.cvtColor(frame_small, cv2.COLOR_BGR2GRAY)
        fg_mask = bg_sub.apply(gray)
        edges = cv2.Canny(gray, 80, 160)

        intensity_mean.append(float(gray.mean()))
        intensity_std.append(float(gray.std()))
        entropy_vals.append(float(shannon_entropy(gray)))
        edge_density.append(float((edges > 0).mean()))
        lap_var.append(float(cv2.Laplacian(gray, cv2.CV_32F).var()))
        foreground_ratio.append(float((fg_mask > 0).mean()))
        crowd_proxy.append(float(foreground_ratio[-1] * (1.0 + edge_density[-1])))

        if prev_gray is not None:
            diff = cv2.absdiff(prev_gray, gray)
            temporal_diff_mean.append(float(diff.mean()))
            temporal_diff_std.append(float(diff.std()))
            flow = cv2.calcOpticalFlowFarneback(prev_gray, gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
            mag, hist = flow_histogram(flow, bins=8)
            flow_mean.append(float(mag.mean()))
            flow_std.append(float(mag.std()))
            flow_q90.append(float(np.quantile(mag, 0.90)))
            flow_max.append(float(mag.max()))
            flow_hist_total += hist

        prev_gray = gray

    flow_hist_total /= max(1, len(frames) - 1)
    abrupt_reference = np.median(temporal_diff_mean) if temporal_diff_mean else 0.0
    abrupt_change_ratio = float(np.mean(np.array(temporal_diff_mean) > abrupt_reference)) if temporal_diff_mean else 0.0

    features = [
        float(np.mean(intensity_mean)) if intensity_mean else 0.0,
        float(np.mean(intensity_std)) if intensity_std else 0.0,
        float(np.mean(entropy_vals)) if entropy_vals else 0.0,
        float(np.mean(edge_density)) if edge_density else 0.0,
        float(np.mean(lap_var)) if lap_var else 0.0,
        float(np.mean(foreground_ratio)) if foreground_ratio else 0.0,
        float(np.mean(crowd_proxy)) if crowd_proxy else 0.0,
        float(np.mean(temporal_diff_mean)) if temporal_diff_mean else 0.0,
        float(np.mean(temporal_diff_std)) if temporal_diff_std else 0.0,
        float(np.mean(flow_mean)) if flow_mean else 0.0,
        float(np.mean(flow_std)) if flow_std else 0.0,
        float(np.mean(flow_q90)) if flow_q90 else 0.0,
        float(np.mean(flow_max)) if flow_max else 0.0,
        float(intensity_mean[-1] - intensity_mean[0]) if len(intensity_mean) >= 2 else 0.0,
        float(np.max(flow_mean)) if flow_mean else 0.0,
        abrupt_change_ratio,
    ] + flow_hist_total.tolist()

    return np.array(features, dtype=np.float32)


def preprocess_frames(frames):
    tensors = []
    for frame in frames:
        frame_resized = cv2.resize(frame, IMG_SIZE, interpolation=cv2.INTER_AREA)
        frame_rgb = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.0
        frame_rgb = (frame_rgb - NORM_MEAN) / NORM_STD
        tensors.append(torch.from_numpy(frame_rgb.transpose(2, 0, 1)))
    return torch.stack(tensors, dim=0).float()


class CrowdSequenceDataset(Dataset):
    def __init__(self, manifest_df, frame_cache, split="train"):
        self.manifest = manifest_df.reset_index(drop=True)
        self.frame_cache = frame_cache
        self.split = split

    def __len__(self):
        return len(self.manifest)

    def __getitem__(self, idx):
        row = self.manifest.iloc[idx]
        frame_indices = json.loads(row["frame_indices"])

        if row["source_type"] == "frames":
            frames = load_frame_sequence(self.frame_cache[row["video_id"]], frame_indices)
        else:
            frames = load_video_sequence(row["source_path"], frame_indices)

        if len(frames) != len(frame_indices):
            if len(frames) == 0:
                frames = [np.zeros((IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.uint8) for _ in frame_indices]
            else:
                while len(frames) < len(frame_indices):
                    frames.append(frames[-1].copy())

        if self.split == "train":
            frames = augment_sequence(frames, stable_seed(f"{row['sequence_id']}_{SEED}"))

        handcrafted = compute_handcrafted_features(frames)
        frame_tensor = preprocess_frames(frames)
        label = torch.tensor(float(row["label"]), dtype=torch.float32)
        tabular = torch.tensor(handcrafted, dtype=torch.float32)

        meta = {
            "sequence_id": row["sequence_id"],
            "video_id": row["video_id"],
            "scene": row["scene"],
            "source_path": row["source_path"],
            "split": row["split"],
            "start_frame": int(row["start_frame"]),
            "end_frame": int(row["end_frame"]),
            "center_frame": int(row["center_frame"]),
        }
        return frame_tensor, tabular, label, meta


class EfficientNetGRUHybrid(nn.Module):
    def __init__(self, tabular_dim):
        super().__init__()
        weights = models.EfficientNet_B0_Weights.DEFAULT
        backbone = models.efficientnet_b0(weights=weights)
        self.backbone = backbone.features
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.cnn_dim = 1280

        for param in self.backbone.parameters():
            param.requires_grad = False
        for block in list(self.backbone.children())[-2:]:
            for param in block.parameters():
                param.requires_grad = True

        self.temporal = nn.GRU(
            input_size=self.cnn_dim,
            hidden_size=256,
            num_layers=1,
            batch_first=True,
            bidirectional=True,
        )
        self.tabular_head = nn.Sequential(
            nn.Linear(tabular_dim, 64),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.LayerNorm(64),
        )
        self.classifier = nn.Sequential(
            nn.Linear(512 + 64, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(256, 1),
        )

    def forward(self, frames, tabular):
        batch_size, time_steps, channels, height, width = frames.shape
        x = frames.reshape(batch_size * time_steps, channels, height, width)
        x = self.backbone(x)
        x = self.pool(x).flatten(1)
        x = x.reshape(batch_size, time_steps, self.cnn_dim)

        seq_output, _ = self.temporal(x)
        seq_embedding = seq_output[:, -1, :]
        tab_embedding = self.tabular_head(tabular)
        logits = self.classifier(torch.cat([seq_embedding, tab_embedding], dim=1)).squeeze(1)
        return logits


def safe_auc(y_true, y_score):
    return float(roc_auc_score(y_true, y_score)) if len(np.unique(y_true)) > 1 else np.nan


def safe_ap(y_true, y_score):
    return float(average_precision_score(y_true, y_score)) if len(np.unique(y_true)) > 1 else np.nan


def contiguous_segments(binary_array):
    segments = []
    start = None
    for idx, value in enumerate(binary_array.astype(int)):
        if value == 1 and start is None:
            start = idx
        elif value == 0 and start is not None:
            segments.append((start, idx - 1))
            start = None
    if start is not None:
        segments.append((start, len(binary_array) - 1))
    return segments


def temporal_iou(seg_a, seg_b):
    left = max(seg_a[0], seg_b[0])
    right = min(seg_a[1], seg_b[1])
    intersection = max(0, right - left + 1)
    union = (seg_a[1] - seg_a[0] + 1) + (seg_b[1] - seg_b[0] + 1) - intersection
    return intersection / union if union > 0 else 0.0


def average_precision_from_pr(recalls, precisions):
    if len(recalls) == 0:
        return 0.0
    mrec = np.concatenate(([0.0], recalls, [1.0]))
    mpre = np.concatenate(([1.0], precisions, [0.0]))
    for idx in range(len(mpre) - 2, -1, -1):
        mpre[idx] = max(mpre[idx], mpre[idx + 1])
    valid = np.where(mrec[1:] != mrec[:-1])[0]
    return float(np.sum((mrec[valid + 1] - mrec[valid]) * mpre[valid + 1]))


def temporal_detection_scores(eval_df, threshold=0.5, iou_threshold=0.5):
    ap_scores = []
    iou_scores = []

    for _, group in eval_df.sort_values(["video_id", "center_frame"]).groupby("video_id"):
        y_true = group["y_true"].to_numpy().astype(int)
        y_score = group["y_score"].to_numpy().astype(float)
        y_pred = (y_score >= threshold).astype(int)

        gt_segments = contiguous_segments(y_true)
        pred_segments = contiguous_segments(y_pred)
        pred_scores = [float(y_score[start : end + 1].mean()) for start, end in pred_segments]

        if gt_segments:
            best_gt_ious = [max([temporal_iou(gt, pred) for pred in pred_segments], default=0.0) for gt in gt_segments]
            iou_scores.append(float(np.mean(best_gt_ious)))

        if not gt_segments:
            continue

        if not pred_segments:
            ap_scores.append(0.0)
            continue

        order = np.argsort(pred_scores)[::-1]
        tp = np.zeros(len(order), dtype=np.float32)
        fp = np.zeros(len(order), dtype=np.float32)
        matched_gt = set()

        for rank, pred_idx in enumerate(order):
            pred_seg = pred_segments[pred_idx]
            best_iou = 0.0
            best_gt_idx = None
            for gt_idx, gt_seg in enumerate(gt_segments):
                if gt_idx in matched_gt:
                    continue
                iou_val = temporal_iou(pred_seg, gt_seg)
                if iou_val > best_iou:
                    best_iou = iou_val
                    best_gt_idx = gt_idx
            if best_iou >= iou_threshold and best_gt_idx is not None:
                tp[rank] = 1.0
                matched_gt.add(best_gt_idx)
            else:
                fp[rank] = 1.0

        cum_tp = np.cumsum(tp)
        cum_fp = np.cumsum(fp)
        precisions = cum_tp / (cum_tp + cum_fp + 1e-9)
        recalls = cum_tp / (len(gt_segments) + 1e-9)
        ap_scores.append(average_precision_from_pr(recalls, precisions))

    mean_ap = float(np.mean(ap_scores)) if ap_scores else np.nan
    mean_iou = float(np.mean(iou_scores)) if iou_scores else np.nan
    return mean_ap, mean_iou


def choose_best_threshold(y_true, y_score):
    if len(np.unique(y_true)) < 2:
        return 0.50
    best_threshold = 0.50
    best_f1 = -1
    for threshold in THRESHOLDS:
        preds = (y_score >= threshold).astype(int)
        score = f1_score(y_true, preds, zero_division=0)
        if score > best_f1:
            best_f1 = score
            best_threshold = float(threshold)
    return best_threshold


def unpack_meta(batch_meta):
    result = {}
    for key, value in batch_meta.items():
        if torch.is_tensor(value):
            result[key] = value.cpu().tolist()
        else:
            result[key] = list(value)
    return result


def run_epoch(model, loader, optimizer, criterion, device, train_mode=True):
    if train_mode:
        model.train()
    else:
        model.eval()

    all_loss = []
    all_probs = []
    all_targets = []

    loop = tqdm(loader, desc="Train" if train_mode else "Eval", leave=False)
    for frames, tabular, labels, _ in loop:
        frames = frames.to(device, non_blocking=True)
        tabular = tabular.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        with torch.set_grad_enabled(train_mode):
            logits = model(frames, tabular)
            loss = criterion(logits, labels)
            if train_mode:
                optimizer.zero_grad(set_to_none=True)
                loss.backward()
                optimizer.step()

        probs = torch.sigmoid(logits).detach().cpu().numpy()
        targets = labels.detach().cpu().numpy()
        all_loss.append(float(loss.item()))
        all_probs.append(probs)
        all_targets.append(targets)

    y_true = np.concatenate(all_targets)
    y_score = np.concatenate(all_probs)
    y_pred = (y_score >= 0.50).astype(int)

    metrics = {
        "loss": float(np.mean(all_loss)),
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "auc_roc": safe_auc(y_true, y_score),
        "auc_pr": safe_ap(y_true, y_score),
    }
    return metrics


@torch.no_grad()
def predict_loader(model, loader, device):
    model.eval()
    rows = []
    for frames, tabular, labels, meta in tqdm(loader, desc="Predict", leave=False):
        frames = frames.to(device, non_blocking=True)
        tabular = tabular.to(device, non_blocking=True)
        logits = model(frames, tabular)
        probs = torch.sigmoid(logits).cpu().numpy()
        labels_np = labels.cpu().numpy()
        tabular_np = tabular.cpu().numpy()
        meta_dict = unpack_meta(meta)

        batch_size = len(probs)
        for idx in range(batch_size):
            row = {
                "sequence_id": meta_dict["sequence_id"][idx],
                "video_id": meta_dict["video_id"][idx],
                "scene": meta_dict["scene"][idx],
                "source_path": meta_dict["source_path"][idx],
                "split": meta_dict["split"][idx],
                "start_frame": int(meta_dict["start_frame"][idx]),
                "end_frame": int(meta_dict["end_frame"][idx]),
                "center_frame": int(meta_dict["center_frame"][idx]),
                "y_true": int(labels_np[idx]),
                "y_score": float(probs[idx]),
            }
            for feat_idx, feat_name in enumerate(HANDCRAFTED_FEATURES):
                row[feat_name] = float(tabular_np[idx, feat_idx])
            rows.append(row)

    pred_df = pd.DataFrame(rows)
    return pred_df


def evaluate_prediction_df(pred_df, split_name, threshold):
    pred_df = pred_df.copy()
    pred_df["y_pred"] = (pred_df["y_score"] >= threshold).astype(int)
    y_true = pred_df["y_true"].to_numpy()
    y_score = pred_df["y_score"].to_numpy()
    y_pred = pred_df["y_pred"].to_numpy()
    map_score, iou_score = temporal_detection_scores(pred_df, threshold=threshold, iou_threshold=0.5)

    return {
        "Split": split_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1 Score": f1_score(y_true, y_pred, zero_division=0),
        "AUC-ROC": safe_auc(y_true, y_score),
        "AUC-PR": safe_ap(y_true, y_score),
        "Frame-level AUC": safe_auc(y_true, y_score),
        "mAP@0.5 (temporal)": map_score,
        "IoU (temporal)": iou_score,
    }


@torch.no_grad()
def benchmark_inference(model, loader, device, max_batches=20):
    model.eval()
    total_samples = 0
    start_time = time.perf_counter()
    for batch_idx, (frames, tabular, _, _) in enumerate(loader):
        if batch_idx >= max_batches:
            break
        frames = frames.to(device, non_blocking=True)
        tabular = tabular.to(device, non_blocking=True)
        _ = model(frames, tabular)
        if device.type == "cuda":
            torch.cuda.synchronize()
        total_samples += frames.size(0)
    elapsed = time.perf_counter() - start_time
    fps = total_samples / elapsed if elapsed > 0 else np.nan
    latency_ms = (elapsed / total_samples) * 1000 if total_samples > 0 else np.nan
    return fps, latency_ms, total_samples, elapsed


def save_plot(fig, filename):
    fig.savefig(OUTPUT_DIR / filename, dpi=180, bbox_inches="tight")


if not DATASET_ROOT.exists():
    raise FileNotFoundError(f"Dataset path not found: {DATASET_ROOT}")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
compute_desc = (
    f"GPU: {torch.cuda.get_device_name(0)} ({torch.cuda.get_device_properties(0).total_memory / (1024 ** 3):.2f} GB VRAM)"
    if device.type == "cuda"
    else "CPU only; pipeline configured to stay practical under ~20 GB RAM"
)
print("Compute:", compute_desc)

clips = collect_clips(DATASET_ROOT)
clips["label"] = clips["label"].fillna(-1).astype(int)
clips["xml_path"] = clips["source_path"].apply(find_xml_for_clip)
unknown_count = int((clips["label"] < 0).sum())
if unknown_count:
    print(f"Dropping {unknown_count} clips with uncertain labels.")
clips = clips[clips["label"] >= 0].copy().reset_index(drop=True)

if clips.empty or clips["label"].nunique() < 2:
    raise RuntimeError("The dataset does not contain enough labeled normal/abnormal clips after discovery.")

train_ids, val_ids, test_ids = make_video_splits(clips)
clips["split"] = np.where(
    clips["video_id"].isin(train_ids),
    "train",
    np.where(clips["video_id"].isin(val_ids), "val", "test"),
)
clips.to_csv(OUTPUT_DIR / "split_manifest.csv", index=False)

sequence_manifest, frame_cache = build_sequence_manifest(clips)
sequence_manifest.to_csv(OUTPUT_DIR / "sequence_manifest.csv", index=False)

print("\nClip distribution:")
display(clips.groupby(["split", "label"]).size().rename("num_clips").reset_index())
print("\nSequence distribution:")
display(sequence_manifest.groupby(["split", "label"]).size().rename("num_sequences").reset_index())

train_manifest = sequence_manifest[sequence_manifest["split"] == "train"].reset_index(drop=True)
val_manifest = sequence_manifest[sequence_manifest["split"] == "val"].reset_index(drop=True)
test_manifest = sequence_manifest[sequence_manifest["split"] == "test"].reset_index(drop=True)

train_ds = CrowdSequenceDataset(train_manifest, frame_cache, split="train")
val_ds = CrowdSequenceDataset(val_manifest, frame_cache, split="val")
test_ds = CrowdSequenceDataset(test_manifest, frame_cache, split="test")

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)

model = EfficientNetGRUHybrid(tabular_dim=len(HANDCRAFTED_FEATURES)).to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

history = []
best_val_auc = -np.inf
best_state = None
patience = 0

for epoch in range(1, NUM_EPOCHS + 1):
    train_metrics = run_epoch(model, train_loader, optimizer, criterion, device, train_mode=True)
    val_metrics = run_epoch(model, val_loader, optimizer, criterion, device, train_mode=False)

    row = {
        "epoch": epoch,
        "train_loss": train_metrics["loss"],
        "train_accuracy": train_metrics["accuracy"],
        "train_precision": train_metrics["precision"],
        "train_recall": train_metrics["recall"],
        "train_f1": train_metrics["f1"],
        "train_auc_roc": train_metrics["auc_roc"],
        "train_auc_pr": train_metrics["auc_pr"],
        "val_loss": val_metrics["loss"],
        "val_accuracy": val_metrics["accuracy"],
        "val_precision": val_metrics["precision"],
        "val_recall": val_metrics["recall"],
        "val_f1": val_metrics["f1"],
        "val_auc_roc": val_metrics["auc_roc"],
        "val_auc_pr": val_metrics["auc_pr"],
    }
    history.append(row)
    print(pd.DataFrame([row]).round(4).to_string(index=False))

    current_val_auc = val_metrics["auc_roc"] if not np.isnan(val_metrics["auc_roc"]) else val_metrics["f1"]
    if current_val_auc > best_val_auc:
        best_val_auc = current_val_auc
        best_state = {k: v.cpu() for k, v in model.state_dict().items()}
        patience = 0
    else:
        patience += 1
        if patience >= EARLY_STOPPING:
            print(f"Early stopping at epoch {epoch}")
            break

history_df = pd.DataFrame(history)
history_df.to_csv(OUTPUT_DIR / "training_history.csv", index=False)

if best_state is not None:
    model.load_state_dict(best_state)

val_pred_df = predict_loader(model, val_loader, device)
test_pred_df = predict_loader(model, test_loader, device)
train_eval_sample_manifest = train_manifest.sample(n=min(120, len(train_manifest)), random_state=SEED).reset_index(drop=True)
train_eval_loader = DataLoader(
    CrowdSequenceDataset(train_eval_sample_manifest, frame_cache, split="val"),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=PIN_MEMORY,
)
train_pred_df = predict_loader(model, train_eval_loader, device)

best_threshold = choose_best_threshold(val_pred_df["y_true"].to_numpy(), val_pred_df["y_score"].to_numpy())
print(f"\nBest validation threshold: {best_threshold:.3f}")

train_metrics = evaluate_prediction_df(train_pred_df, "TrainSample", best_threshold)
val_metrics = evaluate_prediction_df(val_pred_df, "Validation", best_threshold)
test_metrics = evaluate_prediction_df(test_pred_df, "Test", best_threshold)

fps, latency_ms, bench_samples, bench_seconds = benchmark_inference(model, test_loader, device, max_batches=20)
test_metrics["FPS"] = fps
test_metrics["Inference Latency (ms/sequence)"] = latency_ms
test_metrics["Benchmark Samples"] = bench_samples
test_metrics["Benchmark Seconds"] = bench_seconds
test_metrics["Compute"] = compute_desc

metrics_df = pd.DataFrame([train_metrics, val_metrics, test_metrics])
metrics_df.to_csv(OUTPUT_DIR / "metrics_summary.csv", index=False)
display(metrics_df.round(4))

val_pred_df["y_pred"] = (val_pred_df["y_score"] >= best_threshold).astype(int)
test_pred_df["y_pred"] = (test_pred_df["y_score"] >= best_threshold).astype(int)
train_pred_df["y_pred"] = (train_pred_df["y_score"] >= best_threshold).astype(int)

val_pred_df.to_csv(OUTPUT_DIR / "raw_validation_predictions.csv", index=False)
test_pred_df.to_csv(OUTPUT_DIR / "raw_test_predictions.csv", index=False)
train_pred_df.to_csv(OUTPUT_DIR / "raw_train_sample_predictions.csv", index=False)

clip_level_test = (
    test_pred_df.groupby(["video_id", "scene", "source_path"], as_index=False)
    .agg(
        y_true=("y_true", "max"),
        y_score=("y_score", "mean"),
        y_pred=("y_pred", "max"),
        n_sequences=("sequence_id", "count"),
    )
)
clip_level_test.to_csv(OUTPUT_DIR / "clip_level_test_predictions.csv", index=False)

model_path = OUTPUT_DIR / "best_hybrid_efficientnet_gru.pth"
torch.save(model.state_dict(), model_path)

config_payload = {
    "model_name": "EfficientNet-B0 + BiGRU + Handcrafted Features",
    "dataset_root": str(DATASET_ROOT),
    "sequence_length": SEQ_LEN,
    "sequence_stride": SEQ_STRIDE,
    "sequence_sample_step": SEQ_SAMPLE_STEP,
    "image_size": IMG_SIZE,
    "feature_names": HANDCRAFTED_FEATURES,
    "threshold": best_threshold,
    "normalization_mean": NORM_MEAN.tolist(),
    "normalization_std": NORM_STD.tolist(),
    "compute": compute_desc,
}
with open(OUTPUT_DIR / "model_config.json", "w", encoding="utf-8") as fp:
    json.dump(config_payload, fp, indent=2)
with open(OUTPUT_DIR / "class_map.json", "w", encoding="utf-8") as fp:
    json.dump({"0": "normal", "1": "abnormal"}, fp, indent=2)

feature_sample_manifest = train_manifest.sample(n=min(200, len(train_manifest)), random_state=SEED).reset_index(drop=True)
feature_probe_loader = DataLoader(
    CrowdSequenceDataset(feature_sample_manifest, frame_cache, split="val"),
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=PIN_MEMORY,
)
feature_probe_df = predict_loader(model, feature_probe_loader, device)
feature_corr = feature_probe_df[HANDCRAFTED_FEATURES].corr()
feature_corr.to_csv(OUTPUT_DIR / "feature_correlation.csv")

scene_error = (
    test_pred_df.assign(correct=(test_pred_df["y_true"] == test_pred_df["y_pred"]).astype(int))
    .groupby(["scene", "y_true"], as_index=False)
    .agg(
        accuracy=("correct", "mean"),
        avg_score=("y_score", "mean"),
        count=("sequence_id", "count"),
    )
)
scene_error.to_csv(OUTPUT_DIR / "scene_error_summary.csv", index=False)

cm = confusion_matrix(test_pred_df["y_true"], test_pred_df["y_pred"])
pd.DataFrame(cm, index=["true_normal", "true_abnormal"], columns=["pred_normal", "pred_abnormal"]).to_csv(
    OUTPUT_DIR / "confusion_matrix.csv"
)

fig1, axes = plt.subplots(2, 2, figsize=(16, 10))
axes[0, 0].plot(history_df["epoch"], history_df["train_loss"], marker="o", label="Train Loss")
axes[0, 0].plot(history_df["epoch"], history_df["val_loss"], marker="o", label="Val Loss")
axes[0, 0].set_title("Training and Validation Loss")
axes[0, 0].set_xlabel("Epoch")
axes[0, 0].legend()

axes[0, 1].plot(history_df["epoch"], history_df["train_accuracy"], marker="o", label="Train Accuracy")
axes[0, 1].plot(history_df["epoch"], history_df["val_accuracy"], marker="o", label="Val Accuracy")
axes[0, 1].set_title("Training and Validation Accuracy")
axes[0, 1].set_xlabel("Epoch")
axes[0, 1].legend()

axes[1, 0].plot(history_df["epoch"], history_df["val_auc_roc"], marker="o", label="Val AUC-ROC")
axes[1, 0].plot(history_df["epoch"], history_df["val_auc_pr"], marker="o", label="Val AUC-PR")
axes[1, 0].set_title("Validation Ranking Metrics")
axes[1, 0].set_xlabel("Epoch")
axes[1, 0].legend()

axes[1, 1].axis("off")
axes[1, 1].text(0.02, 0.95, metrics_df.round(4).to_string(index=False), family="monospace", va="top")
axes[1, 1].set_title("Metrics Snapshot")

plt.tight_layout()
save_plot(fig1, "training_curves.png")
plt.show()

fig2, axes = plt.subplots(2, 3, figsize=(20, 12))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[0, 0])
axes[0, 0].set_title("Confusion Matrix Heatmap")
axes[0, 0].set_xlabel("Predicted")
axes[0, 0].set_ylabel("True")

if len(test_pred_df["y_true"].unique()) > 1:
    fpr, tpr, _ = roc_curve(test_pred_df["y_true"], test_pred_df["y_score"])
    precision_curve, recall_curve, _ = precision_recall_curve(test_pred_df["y_true"], test_pred_df["y_score"])
    axes[0, 1].plot(fpr, tpr, label=f"AUC={test_metrics['AUC-ROC']:.4f}")
    axes[0, 1].plot([0, 1], [0, 1], linestyle="--", color="gray")
    axes[0, 1].set_title("ROC Curve")
    axes[0, 1].set_xlabel("False Positive Rate")
    axes[0, 1].set_ylabel("True Positive Rate")
    axes[0, 1].legend()

    axes[0, 2].plot(recall_curve, precision_curve, label=f"AP={test_metrics['AUC-PR']:.4f}")
    axes[0, 2].set_title("Precision-Recall Curve")
    axes[0, 2].set_xlabel("Recall")
    axes[0, 2].set_ylabel("Precision")
    axes[0, 2].legend()
else:
    axes[0, 1].text(0.5, 0.5, "ROC unavailable", ha="center", va="center")
    axes[0, 2].text(0.5, 0.5, "PR unavailable", ha="center", va="center")

sns.histplot(data=test_pred_df, x="y_score", hue="y_true", bins=20, kde=True, ax=axes[1, 0], element="step", stat="density")
axes[1, 0].set_title("Prediction Score Distribution")

metric_bar = metrics_df[metrics_df["Split"] == "Test"].melt(
    id_vars="Split",
    value_vars=["Accuracy", "Precision", "Recall", "F1 Score", "AUC-ROC", "AUC-PR", "Frame-level AUC"],
    var_name="Metric",
    value_name="Score",
)
sns.barplot(data=metric_bar, x="Score", y="Metric", palette="viridis", ax=axes[1, 1])
axes[1, 1].set_title("Test Metrics")
axes[1, 1].set_xlim(0, 1.05)

scene_heatmap = (
    test_pred_df.pivot_table(index="scene", columns="y_true", values="y_score", aggfunc="mean")
    .rename(columns={0: "normal", 1: "abnormal"})
    .fillna(0.0)
)
sns.heatmap(scene_heatmap, annot=True, fmt=".3f", cmap="magma", ax=axes[1, 2])
axes[1, 2].set_title("Scene vs Mean Score Heatmap")

plt.tight_layout()
save_plot(fig2, "evaluation_panels.png")
plt.show()

fig3, axes = plt.subplots(1, 2, figsize=(18, 7))
sns.heatmap(feature_corr, cmap="coolwarm", center=0, ax=axes[0])
axes[0].set_title("Handcrafted Feature Correlation Heatmap")

feature_importance_proxy = (
    test_pred_df[HANDCRAFTED_FEATURES + ["y_score"]].corr()["y_score"].drop("y_score").abs().sort_values(ascending=False).head(15)
)
sns.barplot(x=feature_importance_proxy.values, y=feature_importance_proxy.index, palette="rocket", ax=axes[1])
axes[1].set_title("Top Feature Correlations With Score")
axes[1].set_xlabel("|corr(feature, prediction score)|")

plt.tight_layout()
save_plot(fig3, "feature_heatmap_and_rankings.png")
plt.show()

testing_files_manifest = pd.DataFrame(
    [
        {"file": str(model_path), "purpose": "Trained PyTorch model weights for inference/testing"},
        {"file": str(OUTPUT_DIR / "model_config.json"), "purpose": "Model and preprocessing configuration"},
        {"file": str(OUTPUT_DIR / "class_map.json"), "purpose": "Label mapping for outputs"},
        {"file": str(OUTPUT_DIR / "split_manifest.csv"), "purpose": "Video-level train/val/test split manifest"},
        {"file": str(OUTPUT_DIR / "sequence_manifest.csv"), "purpose": "Sequence-level testing/training manifest"},
        {"file": str(OUTPUT_DIR / "metrics_summary.csv"), "purpose": "Final evaluation metrics including accuracy"},
        {"file": str(OUTPUT_DIR / "raw_validation_predictions.csv"), "purpose": "Raw validation predictions"},
        {"file": str(OUTPUT_DIR / "raw_test_predictions.csv"), "purpose": "Raw test predictions for project output"},
        {"file": str(OUTPUT_DIR / "clip_level_test_predictions.csv"), "purpose": "Clip-level aggregated predictions"},
        {"file": str(OUTPUT_DIR / "training_history.csv"), "purpose": "Epoch-wise training logs"},
        {"file": str(OUTPUT_DIR / "feature_correlation.csv"), "purpose": "Raw correlation matrix for handcrafted features"},
        {"file": str(OUTPUT_DIR / "scene_error_summary.csv"), "purpose": "Scene-wise raw error statistics"},
        {"file": str(OUTPUT_DIR / "confusion_matrix.csv"), "purpose": "Raw confusion matrix values"},
        {"file": str(OUTPUT_DIR / "training_curves.png"), "purpose": "Training graphs"},
        {"file": str(OUTPUT_DIR / "evaluation_panels.png"), "purpose": "Evaluation graphs and heatmaps"},
        {"file": str(OUTPUT_DIR / "feature_heatmap_and_rankings.png"), "purpose": "Feature heatmap and rankings"},
    ]
)
testing_files_manifest.to_csv(OUTPUT_DIR / "testing_files_manifest.csv", index=False)

print("\nSaved testing/output files:")
display(testing_files_manifest)
print(f"\nProject output folder: {OUTPUT_DIR}")
````

## Architecture of EfficientNet-BiGRU Hybrid Model
```mermaid
graph TD;
    Frames(Sequence of Frames) --> CNN[EfficientNet-B0 Backbone];
    CNN --> Pool[Adaptive Avg Pool];
    Pool --> RNN[Bidirectional GRU];
    Tabular(Handcrafted Temporal Features) --> TabHead[Tabular Head: Linear + LayerNorm];
    RNN --> Concat[Concatenate Features];
    TabHead --> Concat;
    Concat --> Classifier[Classification Head: Linear + ReLU + Dropout];
    Classifier --> Out(Anomaly Prediction Score);
```


## Video Classification Pipeline Workflow
```mermaid
graph TD;
    A[Raw Video/Frames] --> B[Sample Sequences];
    B --> C[Compute Handcrafted Background Subtraction & Flow];
    B --> D[Normalize & Resize RGB Frames];
    C --> E[PyTorch DataLoader];
    D --> E;
    E --> F[EfficientNet-BiGRU Model Inference];
    F --> G[Threshold Calibration];
    G --> H[Final Anomaly Predictions Output];
```


## Evaluation Metrics Formulas

The model performance is evaluated using standard classification metrics:

- **Accuracy**: $ \frac{TP + TN}{TP + TN + FP + FN} $
- **Precision**: $ \frac{TP}{TP + FP} $
- **Recall (Sensitivity)**: $ \frac{TP}{TP + FN} $
- **F1-Score**: $ 2 \times \frac{Precision \times Recall}{Precision + Recall} $

Where:
- **TP**: True Positives (Correctly identified Abnormal)
- **TN**: True Negatives (Correctly identified Normal)
- **FP**: False Positives (Normal misclassified as Abnormal)
- **FN**: False Negatives (Abnormal misclassified as Normal)

## Algorithm 1: Abnormal Crowd Detection Pipeline
```text
Input: Surveillance Video V, Sequence Length L, Stride S
Output: Sequence of Anomaly Predictions P

1: Extract total frames F from V
2: for i = 0 to length(F) - L step S do
3:     Sequence Seq = F[i : i + L]
4:     Tabular T = compute_handcrafted_features(Seq)
5:     Seq_norm = preprocess_and_normalize(Seq)
6:     
7:     // Model Inference
8:     Spatial_Feat = EfficientNetB0(Seq_norm)
9:     Temporal_Feat = BiGRU(Spatial_Feat)
10:    Tabular_Feat = TabularHead(T)
11:    
12:    Fused = Concatenate(Temporal_Feat, Tabular_Feat)
13:    Score = Sigmoid(Classifier(Fused))
14:    
15:    if Score >= Optimal_Threshold then
16:        P[i] = 1 // Abnormal
17:    else
18:        P[i] = 0 // Normal
19:    end if
20: end for
21: return P
```

## Model 3: EfficientNet + Transformer Hybrid


````python
# --- Model 3: EfficientNet + Transformer Hybrid ---
class EfficientNetTransformerHybrid(nn.Module):
    def __init__(self, tabular_dim):
        super().__init__()
        weights = models.EfficientNet_B0_Weights.DEFAULT
        backbone = models.efficientnet_b0(weights=weights)
        self.backbone = backbone.features
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.cnn_dim = 1280

        # Freeze early layers
        for param in self.backbone.parameters():
            param.requires_grad = False
        for block in list(self.backbone.children())[-2:]:
            for param in block.parameters():
                param.requires_grad = True

        encoder_layer = nn.TransformerEncoderLayer(d_model=self.cnn_dim, nhead=8, batch_first=True, dropout=0.2)
        self.temporal = nn.TransformerEncoder(encoder_layer, num_layers=2)
        
        self.tabular_head = nn.Sequential(
            nn.Linear(tabular_dim, 64),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.LayerNorm(64),
        )
        self.classifier = nn.Sequential(
            nn.Linear(self.cnn_dim + 64, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(256, 1),
        )

    def forward(self, frames, tabular):
        batch_size, time_steps, channels, height, width = frames.shape
        x = frames.reshape(batch_size * time_steps, channels, height, width)
        x = self.backbone(x)
        x = self.pool(x).flatten(1)
        x = x.reshape(batch_size, time_steps, self.cnn_dim)
        
        seq_output = self.temporal(x)
        seq_embedding = seq_output.mean(dim=1) # Mean pooling over time
        tab_embedding = self.tabular_head(tabular)
        logits = self.classifier(torch.cat([seq_embedding, tab_embedding], dim=1)).squeeze(1)
        return logits

print("Training EfficientNet + Transformer Hybrid Model...")
model_trans = EfficientNetTransformerHybrid(tabular_dim=len(HANDCRAFTED_FEATURES)).to(device)
optimizer_trans = torch.optim.AdamW(model_trans.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

history_trans = []
best_val_auc_trans = -np.inf
best_state_trans = None
patience_trans = 0

if 'train_loader' in locals():
    start_time_trans = time.perf_counter()
    for epoch in range(1, NUM_EPOCHS + 1):
        train_metrics = run_epoch(model_trans, train_loader, optimizer_trans, criterion, device, train_mode=True)
        val_metrics = run_epoch(model_trans, val_loader, optimizer_trans, criterion, device, train_mode=False)

        row = {
            "epoch": epoch,
            "train_loss": train_metrics["loss"], "train_accuracy": train_metrics["accuracy"],
            "train_precision": train_metrics["precision"], "train_recall": train_metrics["recall"],
            "train_f1": train_metrics["f1"], "train_auc_roc": train_metrics["auc_roc"], "train_auc_pr": train_metrics["auc_pr"],
            "val_loss": val_metrics["loss"], "val_accuracy": val_metrics["accuracy"],
            "val_precision": val_metrics["precision"], "val_recall": val_metrics["recall"],
            "val_f1": val_metrics["f1"], "val_auc_roc": val_metrics["auc_roc"], "val_auc_pr": val_metrics["auc_pr"],
        }
        history_trans.append(row)
        print(f"[EffNet+Trans] Epoch {epoch}: Val AUC-ROC {val_metrics['auc_roc']:.4f}")

        current_val_auc = val_metrics["auc_roc"] if not np.isnan(val_metrics["auc_roc"]) else val_metrics["f1"]
        if current_val_auc > best_val_auc_trans:
            best_val_auc_trans = current_val_auc
            best_state_trans = {k: v.cpu() for k, v in model_trans.state_dict().items()}
            patience_trans = 0
        else:
            patience_trans += 1
            if patience_trans >= EARLY_STOPPING:
                print(f"Early stopping at epoch {epoch}")
                break

    training_time_trans = time.perf_counter() - start_time_trans
    history_trans_df = pd.DataFrame(history_trans)
    
    if best_state_trans is not None:
        model_trans.load_state_dict(best_state_trans)

    val_pred_df_t = predict_loader(model_trans, val_loader, device)
    test_pred_df_t = predict_loader(model_trans, test_loader, device)
    
    best_threshold_t = choose_best_threshold(val_pred_df_t["y_true"].to_numpy(), val_pred_df_t["y_score"].to_numpy())
    test_metrics_t = evaluate_prediction_df(test_pred_df_t, "Test (EfficientNet+Transformer)", best_threshold_t)
    
    fps_t, latency_ms_t, _, bench_seconds_t = benchmark_inference(model_trans, test_loader, device, max_batches=20)
    test_metrics_t["FPS"] = fps_t
    test_metrics_t["Inference Latency (ms)"] = latency_ms_t
    test_metrics_t["Training Time (s)"] = training_time_trans
    
    metrics_trans_df = pd.DataFrame([test_metrics_t])
    display(metrics_trans_df.round(4))
    
    test_pred_df_t["y_pred"] = (test_pred_df_t["y_score"] >= best_threshold_t).astype(int)
    cm_trans = confusion_matrix(test_pred_df_t["y_true"], test_pred_df_t["y_pred"])
    
    # Eff+Trans Individual Plots
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    axes[0].plot(history_trans_df["epoch"], history_trans_df["train_loss"], marker="o", label="Train Loss")
    axes[0].plot(history_trans_df["epoch"], history_trans_df["val_loss"], marker="o", label="Val Loss")
    axes[0].set_title("EffNet+Trans Loss")
    axes[0].legend()
    
    sns.heatmap(cm_trans, annot=True, fmt="d", cmap="Blues", ax=axes[1])
    axes[1].set_title("EffNet+Trans Confusion Matrix")
    
    if len(test_pred_df_t["y_true"].unique()) > 1:
        fpr, tpr, _ = roc_curve(test_pred_df_t["y_true"], test_pred_df_t["y_score"])
        axes[2].plot(fpr, tpr, label=f"AUC={test_metrics_t['AUC-ROC']:.4f}")
        axes[2].plot([0,1],[0,1], linestyle='--')
        axes[2].set_title("EffNet+Trans ROC")
        axes[2].legend()
    plt.tight_layout()
    plt.show()
````

## Grad-CAM Model Interpretability


````python
import torch.nn.functional as F

class SequenceGradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.feature_maps = None
        self.gradients = None
        
        self.target_layer.register_forward_hook(self.save_feature_maps)
        self.target_layer.register_full_backward_hook(self.save_gradients)

    def save_feature_maps(self, module, input, output):
        self.feature_maps = output

    def save_gradients(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def generate_cam(self, frames, tabular):
        self.model.eval()
        self.model.zero_grad()
        
        logits = self.model(frames, tabular)
        score = torch.sigmoid(logits)[0]
        
        logits.backward()
        
        # average pool gradients spatially
        weights = torch.mean(self.gradients, dim=(2, 3), keepdim=True)
        # weighted sum of feature maps
        cam = torch.sum(weights * self.feature_maps, dim=1, keepdim=True)
        cam = F.relu(cam)
        cam = F.interpolate(cam, size=(IMG_SIZE[1], IMG_SIZE[0]), mode='bilinear', align_corners=False)
        cam = cam.squeeze().cpu().detach().numpy()
        
        cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-8)
        return cam, score.item()

if 'train_loader' in locals():
    print("Generating Grad-CAM for sample...")
    # Get one batch from test set
    test_iter = iter(test_loader)
    frames_batch, tabular_batch, labels_batch, meta_batch = next(test_iter)
    
    frames_single = frames_batch[0:1].to(device)
    tabular_single = tabular_batch[0:1].to(device)
    label_single = labels_batch[0:1].item()
    
    # Target the last CNN layer of the original backbone (EfficientNet-B0)
    target_layer_gru = model.backbone[-1]
    gradcam_gru = SequenceGradCAM(model, target_layer_gru)
    
    cam, score = gradcam_gru.generate_cam(frames_single, tabular_single)
    
    center_idx = frames_single.size(1) // 2
    raw_frame = frames_single[0, center_idx].cpu().numpy().transpose(1, 2, 0)
    raw_frame = (raw_frame * NORM_STD + NORM_MEAN)
    raw_frame = np.clip(raw_frame, 0, 1)
    
    # CAM is for the sequence
    cam_center = cam[center_idx] if len(cam.shape) == 3 else cam
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(raw_frame)
    axes[0].set_title(f"Original Center Frame (True: {label_single})")
    axes[0].axis('off')
    
    heatmap = cv2.applyColorMap(np.uint8(255 * cam_center), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB) / 255.0
    axes[1].imshow(heatmap)
    axes[1].set_title(f"Grad-CAM Heatmap")
    axes[1].axis('off')
    
    overlay = 0.5 * raw_frame + 0.5 * heatmap
    axes[2].imshow(overlay)
    axes[2].set_title(f"Overlay (Pred Score: {score:.3f})")
    axes[2].axis('off')
    
    plt.tight_layout()
    plt.show()
````
